### Extraccion de los géneros de peliculas como vector One-Hot

```python
import pandas as pd
```

```python
pelis=pd.read_csv("/Users/antonio/OneDrive - farmasturias.org/doctorado/paper_generos_peliculas/XAI_movies/cjtos/ml-25M/movies.csv")
```

```python
pelis
```

Los generos vienen en la columna genres separados por una barra.
genero primero un conjunto con las distintas variaciones

```python
cto_generos=set()
generos=set()
cto_generos=set(pelis["genres"])
```

```python
len(cto_generos)
```

recorremos todas las combinaciones y añadimos los valores a un conjunto (no admite duplicados)

```python
for e in cto_generos:
    spl=e.split(sep='|')
    for g in spl:
        generos.add(str(g)) 
```

```python
datageneros=pd.DataFrame(list(generos))
```

```python
datageneros
```

la funcion get_dummies nos genera la codificación one-hot

```python
onehot=pd.get_dummies(datageneros)
```

```python
onehot
```

```python
final=datageneros.merge(onehot,how='inner', left_index=True, right_index=True)
```

```python
final=final.rename(columns={0:'literal'})
final
```

 puesto que cada pelicula puede tener mas un genero sumamos los vectores one-hot de cada genero de la misma
 para ello recorremos las 62,000 peliculas

#### Iteramos todas las peliculas

```python
a=0
result=pd.DataFrame()
for index,row in pelis.iterrows():
    a+=1
    if a%10000==0:
        print(a)
    dataframe=pd.DataFrame()
    genrer_peli=row["genres"]
    gen=genrer_peli.split(sep='|')
    for g in gen:
        match=final[final["literal"]==g]
        dataframe=dataframe.append(match)
    serie=dataframe[:][1:].sum()
    valoracion=pd.Series(row["title"]).append(serie) # ver si se puede eliminar todo esto del rename
    valoracion=valoracion.rename(index={0:'title'})
    valoracion=pd.Series(row["movieId"]).append(valoracion)
    result=result.append(valoracion,ignore_index=True)
    
    
```

```python
result=result.rename(columns={0:'item_id'})

result
```

Exportamos a csv, la opcion index=false elimina de la exportacion los nombres de las filas

```python
result.to_csv( "/Users/antonio/OneDrive - farmasturias.org/doctorado/paper_generos_peliculas/XAI_movies/cjtos/ml-25M/movies_one_hot_genre.csv",index=False)
```
